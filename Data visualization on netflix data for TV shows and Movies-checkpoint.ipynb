# Year wise number of Tv shows by bar graph:-

```python
dnt_tvshows = dnt[dnt['type'] == 'TV Show']
dnt_tvshows["release_year"].value_counts()
```

```python
plt.figure(figsize  = (15,8))
dnt_tvshows["release_year"].value_counts().plot(kind = 'bar',xlabel = 'year',ylabel = 'no. of TV Shows',color = 'Red',title = "TV Shows per year")
```

# Year wise number of movies by bar graph:-

# Items in each NetFlix rating:-

# Top 10 directors of Movies based on number of titles:-

```python
plt.figure(figsize = (9,8))
dnt_tvm = dnt[dnt['type']=='Movie']
dnt_tvm.director.value_counts().head(10).plot(kind = 'bar',xlabel ='Director' ,ylabel = 'no.of shows',color = 'Grey',title = 'Top 10  Directors')
print('The country of highest Movies \n',dnt_tvm.director.value_counts().head(10))
```

# Top 10 Actors on Netflix:-

```python
plt.figure(figsize = (15,15))
plt.title("Top 10 Actors Based on The Number of Titles")
filtr_cast = dnt[dnt.cast!="No Cast"].set_index("title").cast.str.split(' , ' , expand = True).stack().reset_index(level=1, drop = True)
sns.countplot(y = filtr_cast , order = filtr_cast.value_counts().index[:10])
plt.show()
```

# Top 10 Countries as per Number of Shows

```python
dnt.fillna({"country":"Not Available"})
plt.figure(figsize = (15,15))
plt.title("Top 10 Countries Based on The Number of Shows")
filtr_ctr = dnt[dnt.country!= "Not Available"].set_index("title").country.str.split(' , ' , expand = True).stack().reset_index(level=1, drop = True)
sns.countplot(y = filtr_ctr , order = filtr_ctr.value_counts().index[:10])
plt.show()
```
